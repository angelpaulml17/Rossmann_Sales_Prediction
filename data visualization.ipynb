```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
store = pd.read_csv('./store (2).csv')
train = pd.read_csv('./train(2) (2).csv')
test = pd.read_csv('./test(1) (2).csv')
```

```python
# Univariate Analysis
# Histograms
store.hist(bins=50, figsize=(15, 10))
plt.suptitle('Histograms of All Variables')
plt.show()
```

```python
# Box Plots
store.plot(kind='box', subplots=True, layout=(3,3), sharex=False, sharey=False, figsize=(15, 10))
plt.suptitle('Box Plots of All Variables')
plt.show()
```

```python
# Density Plots
store.plot(kind='density', subplots=True, layout=(3,3), sharex=False, figsize=(15, 10))
plt.suptitle('Density Plots of All Variables')
plt.show()
```

```python
# Multivariate Analysis
# Scatter Plots
sns.pairplot(store)
plt.suptitle('Pairwise Scatter Plots')
plt.show()
```

```python
# Correlation Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(store.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Univariate Analysis
# Histograms
train.hist(bins=50, figsize=(20, 15))
plt.suptitle('Histograms of All Variables')
plt.show()
```

```python
# Multivariate Analysis
# Scatter Plots
sns.pairplot(train)
plt.suptitle('Pairwise Scatter Plots')
plt.show()
```

```python
# Correlation Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(train.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Counting the number of stores for each StoreType
store_count_by_storetype = store['StoreType'].value_counts()

# Counting the number of stores for each Assortment type
store_count_by_assortment = store['Assortment'].value_counts()

# Plotting count of stores by StoreType
plt.figure(figsize=(10, 6))
sns.barplot(x=store_count_by_storetype.index, y=store_count_by_storetype.values)
plt.title('Count of Stores per Store Type')
plt.xlabel('Store Type')
plt.ylabel('Number of Stores')
plt.show()

# Plotting count of stores by Assortment
plt.figure(figsize=(10, 6))
sns.barplot(x=store_count_by_assortment.index, y=store_count_by_assortment.values)
plt.title('Count of Stores per Assortment Type')
plt.xlabel('Assortment Type')
plt.ylabel('Number of Stores')
plt.show()
```

```python
# Group the data by day of the week and calculate total sales for each day
daily_sales = train.groupby(train['DayOfWeek'])['Sales'].sum()

# Define labels for the days of the week
days_of_week = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

# Create a bar plot for sales by day of the week
plt.figure(figsize=(10, 6))
plt.bar(days_of_week, daily_sales)
plt.xlabel('Day of the Week')
plt.ylabel('Total Sales')
plt.title('Total Sales by Day of the Week')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(True)
plt.show()
```

```python

# Binning competition distances into categories
bins = [0, 1000, 2000, 3000, 4000, 5000, 10000, 20000, float('inf')]
labels = ['0-1000', '1001-2000', '2001-3000', '3001-4000', '4001-5000', '5001-10000', '10001-20000', '20001+']
train['CompetitionDistanceGroup'] = pd.cut(train['CompetitionDistance'], bins=bins, labels=labels)

# Calculate average sales for each competition distance category
avg_sales_by_distance = train.groupby('CompetitionDistanceGroup')['Sales'].mean()

# Create a bar plot for sales based on competition distance
plt.figure(figsize=(10, 6))
avg_sales_by_distance.plot(kind='bar', color='skyblue')
plt.xlabel('Competition Distance Group')
plt.ylabel('Average Sales')
plt.title('Average Sales Based on Competition Distance')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
